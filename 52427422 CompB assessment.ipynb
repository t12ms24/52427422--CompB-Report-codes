```python
import os 
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
from scipy import stats
```

```python
my_data_directory = "C:\\Users\\user\\OneDrive - University of Aberdeen"
all_data = []

for filename in os.listdir(my_data_directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(my_data_directory, filename)
        df = pd.read_csv(file_path, sep="\t")
        all_data.append(df)

data = pd.concat(all_data, ignore_index=True) 
data
```

```python
analysis_results = []
for index, gene_row in data.iterrows():
    Gene=gene_row['Gene']
    Wildtype_Sequence=gene_row["WildType.Sequence"]
    Mutant_Sequence=gene_row["Mutant.Sequence"]

    wt_viability=[gene_row["CellViability.WT.Rep1"], gene_row["CellViability.WT.Rep2"], gene_row["CellViability.WT.Rep3"]]
    mut_viability=[gene_row["CellViability.Mut.Rep1"], gene_row["CellViability.Mut.Rep2"], gene_row["CellViability.Mut.Rep3"]]
    
    wt_mrna=[gene_row["mRNA.Expression.WT.Rep1"], gene_row["mRNA.Expression.WT.Rep2"], gene_row["mRNA.Expression.WT.Rep3"]]
    mut_mrna=[gene_row["mRNA.Expression.Mut.Rep1"], gene_row["mRNA.Expression.Mut.Rep2"], gene_row["mRNA.Expression.Mut.Rep3"]]
    
    wt_protein=[gene_row["Protein.Expression.WT.Rep1"], gene_row["Protein.Expression.WT.Rep2"], gene_row["Protein.Expression.WT.Rep3"]]
    mut_protein=[gene_row["Protein.Expression.Mut.Rep1"], gene_row["Protein.Expression.Mut.Rep2"], gene_row["Protein.Expression.Mut.Rep3"]]
    
    if len(Wildtype_Sequence)==len(Mutant_Sequence):
        mutation_type="substitution"
    elif len(Wildtype_Sequence)<len(Mutant_Sequence):
        mutation_type="insertion"
    else:
        mutation_type="deletion"
    
    promoter_length=1000
    first_diff_position=None
    for i in range(min(len(Wildtype_Sequence), len(Mutant_Sequence))):
        if Wildtype_Sequence[i]!=Mutant_Sequence[i]:
            first_diff_position=i
            break
    if first_diff_position is None and len(Wildtype_Sequence)!=len(Mutant_Sequence):
        first_diff_position=min(len(Wildtype_Sequence), len(Mutant_Sequence))

    mutation_location="promoter" if first_diff_position is not None and first_diff_position<promoter_length else "coding"

    # the averages of the three replicates for each
    wt_viability_avg=sum(wt_viability)/3
    mut_viability_avg=sum(mut_viability)/3
    wt_mrna_avg=sum(wt_mrna)/3
    mut_mrna_avg=sum(mut_mrna)/3
    wt_protein_avg=sum(wt_protein)/3
    mut_protein_avg=sum(mut_protein)/3

    # the percentage changes
    viability_change=((mut_viability_avg-wt_viability_avg)/wt_viability_avg) * 100
    mrna_change=((mut_mrna_avg-wt_mrna_avg)/wt_mrna_avg) * 100
    protein_change=((mut_protein_avg-wt_protein_avg)/wt_protein_avg) * 100

    # determining the statistical significance in the changes of cell viability and how differences in protein and mrna expression differ
    viability_significant=stats.ttest_ind(wt_viability, mut_viability).pvalue < 0.05
    mrna_significant=stats.ttest_ind(wt_mrna, mut_mrna).pvalue < 0.05
    protein_significant=stats.ttest_ind(wt_protein, mut_protein).pvalue < 0.05

    # Save result for this gene
    analysis_results.append({
        "Gene": Gene,
        "MutationType": mutation_type,
        "MutationLocation": mutation_location,
        "FirstDiffPosition": first_diff_position,
        "ViabilityChange": viability_change,
        "ViabilitySignificant": viability_significant,
        "MRNAChange": mrna_change,
        "MRNASignificant": mrna_significant,
        "ProteinChange": protein_change,
        "ProteinSignificant": protein_significant
    })

summary_df=pd.DataFrame(analysis_results)
```

```python
summary_df["TotalImpact"]=(
    summary_df["ViabilityChange"]+ 
    summary_df["MRNAChange"] + 
    summary_df["ProteinChange"] 
)

top5_genes_df = summary_df.sort_values(by='TotalImpact',ascending=False).head(5)
top5_genes_df
```

```python
fig, ax = plt.subplot_mosaic([["box", "scatter"],["hist", "hist"]],
                             figsize=(20, 15), layout="constrained")
sbn.boxplot(top5_genes_df, x="MutationType", y="ViabilityChange", ax=ax["box"])
ax["box"].set_title("Effect of Mutation Type on Cell Viability")
ax["box"].set_xlabel("Mutation Type")
ax["box"].set_ylabel("% Cell Viability Change")
sbn.scatterplot(top5_genes_df, x="Gene", y="TotalImpact", hue="ViabilityChange", ax=ax["scatter"]) 
ax["scatter"].set_title("Impact of gene mutations on cell viability")
ax["scatter"].set_xlabel("Gene")
ax["scatter"].set_ylabel("TotalImpact")
sbn.kdeplot(top5_genes_df, x="ViabilityChange", hue="MutationLocation", ax=ax["hist"])
ax["hist"].set_title("Distribution of Cell Viability by Mutation Location")
ax["hist"].set_xlabel("% Viability Change")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sbn.scatterplot(data=top5_genes_df, x="Gene", y="MRNAChange",hue="MutationType")
plt.title("Effects of mutation type on mRNA excpression change")
plt.xlabel("Gene")
plt.ylabel("% MRNAChange")
plt.tight_layout()
plt.show() 
```

```python
plt.figure(figsize=(10, 6))
sbn.barplot(data=top5_genes_df, x="Gene", y="ProteinChange",hue="MutationType")
plt.title("Protein expression changes by Mutation Type")
plt.xlabel("Gene")
plt.ylabel("% ProteinChange")
plt.tight_layout()
plt.show() 
```
